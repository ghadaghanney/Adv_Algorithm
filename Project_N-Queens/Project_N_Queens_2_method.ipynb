```python
class echiquier_n_reines:
    def __init__(self, taille):
        # tableau a des dimensions taille x taille
        self.taille = taille
        # colonnes [r] est un nombre c si une reine est placée à la ligne r et à la colonne c.
        # les colonnes [r] sont hors limites si aucune reine n'est placée dans la ligne r.
        # une fois toutes les reines placées, elles seront à des positions
        # (colonnes [0], 0), (colonnes [1], 1), ... (colonnes [taille - 1], taille - 1)
        self.colonnes = []
 
    def get_taille(self):
        return self.taille
 
    def get_compter_reines(self):
        return len(self.colonnes)
 
    def place_ligne_suivant(self, c):
        self.colonnes.append(c)
 
    def supprimer_ligne_courant(self):
        return self.colonnes.pop()
 
    def colonne_sécurise_ligne_suivant(self, c):
        # index de la ligne suivante
        l = len(self.colonnes)
 
        # vérifier la colonne
        for reine_colonne in self.colonnes:
            if c == reine_colonne:
                return False
 
        # vérifier la diagonale
        for reine_ligne, reine_colonne in enumerate(self.colonnes):
            if reine_colonne - reine_ligne == c - l:
                return False
 
        # vérifier une autre diagonale
        for reine_ligne, reine_colonne in enumerate(self.colonnes):
            if ((self.taille - reine_colonne) - reine_ligne
                == (self.taille - c) - l):
                return False
 
        return True
 
    def placer_reine(self):
        for l in range(self.taille):
            for c in range(self.taille):
                if c == self.colonnes[l]:
                    print('Q', end=' ')
                else:
                    print('1', end=' ')
            print()
 

def placer_toutes_reines(taille):
    Echiquier = echiquier_n_reines(taille) #Afficher un échiquier pour chaque configuration possible de placement 
                                           #de n reines sur un échiquier n x n où n = taille et 
                                           # imprimez le nombre de ces configurations
    nombre_des_solutions = toute_solution(Echiquier)
    print('Nombre des solutions:', nombre_des_solutions)
 
def toute_solution(Echiquier):
    #Afficher un échiquier pour chaque configuration possible de remplissage du
    # Echiquier donné avec des reines et renvoyer le nombre de ces configurations.
    taille = Echiquier.get_taille()
 
    # si l'echiquier est plein , afficher la solution placer_reine
    if taille == Echiquier.get_compter_reines():
        Echiquier.placer_reine()
        print()
        return 1
 
    nombre_des_solutions = 0
    # placer la reine dans la ligne suivante
    for c in range(taille):
        if Echiquier.colonne_sécurise_ligne_suivant(c):
            Echiquier.place_ligne_suivant(c)
            nombre_des_solutions += toute_solution(Echiquier)
            Echiquier.supprimer_ligne_courant()
 
    return nombre_des_solutions
 
 
n = int(input('Enter n: '))
placer_toutes_reines(n)
```
